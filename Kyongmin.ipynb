```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/kc_house_data.csv')
```

```python
fig, ax = plt.subplots(figsize = (15, 9))
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(abs(df.corr()), mask=mask,annot=True)
```

```python
#Looking at the graph below, we can see that the points are focused on the left hand side. This would be a good 
#time to check the logarithmic transformation. This should make it look normal. (skewed right)
plt.hist(df['price'], bins = (round(len(df)**.5)));
```

```python
#Example of good use for logarithm. 
plt.hist(np.log(df['price']), bins = (round(len(df)**.5)));
```
